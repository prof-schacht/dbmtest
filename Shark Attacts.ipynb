```python
import numpy as np
import pandas as pd

```

```python
df = pd.read_csv('data/attacks.csv', encoding='ISO-8859-1',)
```

```python
df.rename(columns={'Fatal (Y/N)': 'Fatal', 'Sex ': 'Sex'}, inplace=True)
```

```python
df['Fatal'].unique()
```

```python
df.loc[~df['Fatal'].isin(['Y', 'N', 'UNKNOWN'])]
```

```python
df['Fatal'] = df['Fatal'].map(lambda x: str(x).strip().upper()).replace('F', 'Y').map(lambda s: s if s in ('Y', 'N') else np.nan)
```

```python
df['Fatal'].unique()
```

```python
df['Sex'].unique()
```

```python
df['Sex'] = df['Sex'].map(lambda x: str(x).strip().upper()).replace('N', 'M').map(lambda s: s if s in ('F', 'M') else np.nan)
```

```python
table = pd.pivot_table(df, columns='Fatal', values='Date', index='Sex', aggfunc='count')
table.div(table.sum(axis=1), axis=0)
```
